```python
fig, axs = plt.subplots(3, 3, figsize=(15, 12))  # Increased width to ensure more space for y-labels

letters = ['(a)', '(b)', '(c)', '(d)', '(e)', '(f)', '(g)', '(h)', '(i)']
# Iterate over the result_collection to plot each model's data
for ax, (model_name, model_stats_df), letter in zip(axs.flatten(), result_collection.items(), letters):
    melted_df = model_stats_df.melt(id_vars='Train data', var_name='Metric', value_name='Value')
    
    colors = mpl.colormaps['Paired'].colors
    
    for i, metric in enumerate(melted_df['Metric'].unique()):
        subset = melted_df[melted_df['Metric'] == metric]
        for j, train_data in enumerate(subset['Train data'].unique()):
            subset_td = subset[subset['Train data'] == train_data]
            ax.bar(i + j*0.25 - 0.25, subset_td['Value'].values[0], width=0.25, color=colors[j], label=train_data if i == 0 else "", 
                   edgecolor='black', linewidth=0.4)

    ax.set_xticks(range(len(melted_df['Metric'].unique())))
    ax.set_xticklabels(melted_df['Metric'].unique())
    ax.legend(title='Train data', loc='lower right', shadow=True, fontsize='small')
    ax.set_ylabel('Performance score')
    ax.yaxis.set_label_coords(-0.08, 0.5)  # Manually set the label coordinates closer to the axis

    ax.text(0.5, -0.15, letter + " " + model_name, transform=ax.transAxes, size=12, ha='center')

# Increase the space between subplots
plt.subplots_adjust(left=0.15, right=1.95, top=0.95, bottom=0.05, hspace=0.5, wspace=0.5)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.savefig('performance_comparison_by_train_data.pdf', bbox_inches='tight')
plt.show()
```
